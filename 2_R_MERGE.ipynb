```r
# Crear vectores con el número de goles de cada jugador por temporada
estebanGoles <- c(15,12,14,20)
lucasGoles <-   c(7,9,12,4)
eduardoGoles <-c(2,4,5,6)
alexisGoles <- c(12,22,31,15)
```

```r
generarGraficoEjemplo <- function(datos, legendaDatos, filas, columnas) {
  # Graficar los datos seleccionados
  matplot(t(datos[c(filas), c(columnas)]), type = "b", pch = 15:20, col = c(1:4))
  # Añadir leyenda
  legend("topright", inset = 0.01, legend = legendaDatos, pch = 15:20, col = c(1:4))
}
```

```r
# Definir una función para generar gráficos personalizados
# `datos`: matriz de datos a graficar
# `legendaDatos`: vector con los nombres de los jugadores
# `filas`: índices de las filas (jugadores) a graficar
# `columnas`: índices de las columnas (temporadas) a graficar
generarGraficoEjemplo <- function(datos, legendaDatos, filas, columnas) {
  # Graficar los datos seleccionados
  matplot(t(datos[c(filas), c(columnas)]), type = "b", pch = 15:20, col = c(1:4))
  # Añadir leyenda
  legend("topright", inset = 0.01, legend = legendaDatos, pch = 15:20, col = c(1:4))
}
```

```r
# Definir la función con legendaDatos como argumento
generarGraficoEjemplo <- function(datos, legendaDatos, filas, columnas) {
  # Graficar los datos seleccionados
  matplot(t(datos[c(filas), c(columnas)]), type = "b", pch = 15:20, col = c(1:4))
  # Añadir leyenda
  legend("topleft", inset = 0.01, legend = legendaDatos[filas], pch = 15:20, col = c(1:4))
}

# Llamar a la función para generar un gráfico con todos los jugadores y todas las temporadas
generarGraficoEjemplo(matrizGoles, jugador, 1:4, 1:4)

# Imprimir la matriz de goles una vez más
matrizGoles
```
