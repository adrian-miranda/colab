```r

# summary() para obtener un resumen detallado de los resultados de un modelo estadístico:
# minimo maximo mediana quintiles 1 .25 y 3 .75 y mean , promedio quizas
summary(mpg$cty) #Rendimientos en millas x galon en ciudad
```

```r
#Quintiles
#El [25%, 50%, 75%] de los valores de cty SON MENORES o iguales a X
quantile(mpg$cty,probs=c(0.25,0.5,0.75))
```

```r
#mean= media aritmética promedio de todos los valores
mean(mpg$cty)
```

```r
#median valor del medio en un conjunto de datos
median(mpg$cty)
```

```r
sqrt(var(mpg$cty)) #Desviación estandar
```

```r
sd(mpg$cty)  #Desviación estandar
```

```r
sd(mpg$hwy) #millas por galon en carretera
```

```r
#Coeficiente de variacion
#Se usa para evaluar dispersión relativa de una distribución de datos, a diferencia de SD, esto mide las dispesion en una misma unidad
#se expresa en %, permite comparar la variabilidad de los conjuntos de datons  qe tienen distintas escalas o unidades
sd(mpg$cty)/mean(mpg$cty)
sd(mpg$hwy)/mean(mpg$hwy)
```

```r
#Correlación de variables
#busca cuantificar el nivel de asociación que existe entre 2 variables
#coeficiente de correlación entre dos variables, el mas popular es correlación de pearson
#valores posibles -1 y 1
#1 se dice que es una correlación positiva o directamente proporcional, con pendiente positiva de la gráfica
#-1 negativa e inversamente proporcional
#0 es neutra
```

```r
library(dplyr) #libreria de manipulación de tablas
```

```r
datos<-select(mpg,displ,hwy)  #hwy es millas por galon y displ es cilindrada
datos
```

```r
ggplot(datos)+aes(x=displ,y=hwy)+geom_point()+geom_smooth(method = "lm")
```

```r
ggplot(datos %>% filter(displ<5))+aes(x=displ,y=hwy)+geom_point()+geom_smooth(method = "auto")
```

```r
cor(datos) #pearson
```

```r
cor(datos,method="spearman")
```

```r
cor(datos,method="kendall")
```
